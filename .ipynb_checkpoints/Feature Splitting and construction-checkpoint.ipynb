```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df = pd.read_csv(r"E:\Jupyter Notebooks\train.csv")[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
x = df.drop('Survived',axis=1)
y = df.loc[:,'Survived']
```

```python
x.shape, y.shape
```

```python
np.mean(cross_val_score(LogisticRegression(), x, y, scoring = 'accuracy'))
```

### Feature Construction

```python
x['family_size'] = x['SibSp'] + x['Parch'] + 1 # adding 1 for counting the person himself
```

```python
x
```

```python
def fun(col):
    if(col == 1):
        return 0 #alone
    elif(col > 1 and col<=4):
        return 1 #small family
    else:
        return 2 # large family
```

```python
x['family_type'] = x['family_size'].apply(fun)
```

```python
x
```

```python
x.drop(columns=['SibSp','Parch','family_size'],inplace=True)
```

```python
x
```

```python
np.mean(cross_val_score(LogisticRegression(), x, y, scoring = 'accuracy'))
```
