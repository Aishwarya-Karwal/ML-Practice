```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import FunctionTransformer
```

```python
data = pd.read_csv(r"E:\Jupyter Notebooks\train.csv")
```

```python
data.head()
```

```python
data.drop(['PassengerId','Pclass','Name','Sex','SibSp','Parch','Ticket','Cabin','Embarked'], axis = 1, inplace = True)
```

```python
data.sample()
```

```python
# there should be no null values to apply transformations 
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].mean(), inplace = True)
#data['Age'].mean()
```

```python
data.Age
```

## Doing Train test split

```python
xtrain, xtest, ytrain, ytest = train_test_split(data.drop(['Survived'], axis = 1), 
                                                data.iloc[:,0], random_state=1
                                               , test_size = 0.2)
```

```python
xtrain
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
ytrain.head()
```

```python
ytrain.head()
```

```python
## Now checking the distributions of the columns
for col in xtrain.columns:
    plt.figure(figsize = (14,5))
    plt.subplot(121)
    sns.distplot(xtrain[col])
    plt.title("{0} pdf".format(col))
    
    plt.subplot(122)
    stats.probplot(xtrain[col], fit= True, dist='norm', plot = plt)
    plt.title("{0} QQ plot".format(col))
    
    plt.show()
```

```python
# from the above plots we can say that age is somewhat normal , but fare is right skewed
```

## Plotting models without function transform

### checking the impact, on two diff classifiers , logistic regression will get affected with normality of the data while decision treeClassifier would not 

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(xtrain, ytrain)
clf2.fit(xtrain, ytrain)

pred = clf1.predict(xtest)
predTree = clf2.predict(xtest)
```

```python
accuracy_score(ytest, pred),  accuracy_score(ytest, predTree)
```

```python
print(np.mean(cross_val_score(clf1, xtrain, ytrain, scoring = 'accuracy', cv = 10)), " logistic regression") 
print(np.mean(cross_val_score(clf2, xtrain, ytrain, scoring = 'accuracy', cv = 10)), " tree classifier") 
```

```python
# don't know why logistic is doing better than tree and is this a general trend
```

## modeling after function transformer

```python
ft = FunctionTransformer(func = np.log1p)

xtrainTransformed = ft.fit_transform(xtrain)
xtestTransformed = ft.transform(xtest)
```

```python
xtrainTransformed.shape, xtestTransformed.shape
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
clf1.fit(xtrainTransformed, ytrain)
clf2.fit(xtrainTransformed, ytrain)

predT = clf1.predict(xtestTransformed)
predTreeT = clf2.predict(xtestTransformed)
```

```python
accuracy_score(ytest, predT),  accuracy_score(ytest, predTreeT)
```

```python
print(np.mean(cross_val_score(clf1, xtrainTransformed, ytrain, scoring = 'accuracy', cv = 10)), " logistic regression") 
print(np.mean(cross_val_score(clf2, xtrainTransformed, ytrain, scoring = 'accuracy', cv = 10)), " tree classifier")
```

```python
## Now checking the distributions of the transformed columns
for col in xtrainTransformed.columns:
    plt.figure(figsize = (14,5))
    plt.subplot(121)
    stats.probplot(xtrain[col], dist = "norm", plot=plt)
    plt.title("{0} before transform".format(col))
    
    plt.subplot(122)
    stats.probplot(xtrainTransformed[col], dist = "norm", plot=plt)
    plt.title("{0} after transform".format(col))
    plt.show()
```

```python
# Age was better before while Fare is better normally distributed afterwards.
```

## Creating a function to apply custom transforms 

```python
def applyTransform(transform):
    tf = FunctionTransformer(func = transform)
    xtrainT = tf.fit_transform(xtrain)
    xtestT = tf.transform(xtest)
    
    # model
    clf = LogisticRegression()
    clf.fit(xtrainT, ytrain)
    pred = clf.predict(xtestT)
    
    # metric - after cross val
    print("Accuracy = ", np.mean(cross_val_score(clf, xtrainT, ytrain, scoring = 'accuracy', cv = 10)))
    
    # plot
    plt.figure(figsize = (14,5))
    plt.subplot(121)
    stats.probplot(xtrain["Fare"], dist = "norm", plot=plt)
    plt.title("Fare before transform")
    
    plt.subplot(122)
    stats.probplot(xtrainTransformed["Fare"], dist = "norm", plot=plt)
    plt.title("Fare after transform")
    plt.show()
```

```python
applyTransform(lambda x : x**1/2) # squareroot transform
```
