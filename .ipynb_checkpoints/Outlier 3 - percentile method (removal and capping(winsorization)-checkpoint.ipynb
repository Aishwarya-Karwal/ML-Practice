Winsorizing or winsorization is the transformation of statistics by limiting extreme values in the statistical data to reduce the effect of possibly spurious outliers

in the case of percentile capping its called winsorising as told in the video 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"E:\Jupyter Notebooks\placement.csv")
```

```python
df.isnull().sum()
```

```python
df.sample(5)
```

```python
upperlimit = df['placement_exam_marks'].quantile(0.99)
lowerlimit =  df['placement_exam_marks'].quantile(0.01)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.distplot(df['placement_exam_marks'])
    
plt.subplot(1,2,2)
sns.boxplot(df['placement_exam_marks'])
```

```python
# trimming 
newdf = df[(df['placement_exam_marks'] >= lowerlimit) & (df['placement_exam_marks'] <= upperlimit)]
```

```python
# capping 
newdfcap = df.copy()
newdfcap['placement_exam_marks'] = np.where(
newdfcap['placement_exam_marks'] > upperlimit, upperlimit ,
np.where(newdfcap['placement_exam_marks'] < lowerlimit, lowerlimit, newdfcap['placement_exam_marks']))
```

```python
plt.figure(figsize=(17,15))
plt.subplot(3,2,1)
sns.distplot(df['placement_exam_marks'])
    
plt.subplot(3,2,2)
sns.boxplot(df['placement_exam_marks'])

plt.subplot(3,2,3)
sns.distplot(newdf['placement_exam_marks'])
    
plt.subplot(3,2,4)
sns.boxplot(newdf['placement_exam_marks'])

plt.subplot(3,2,5)
sns.distplot(newdfcap['placement_exam_marks'])

plt.subplot(3,2,6)
sns.boxplot(newdfcap['placement_exam_marks'])
```

```python
# for cgpa column
```

```python
upperlimit = df['cgpa'].quantile(0.99)
lowerlimit =  df['cgpa'].quantile(0.01)

newdf2 = df[(df['cgpa'] >= lowerlimit) & (df['cgpa'] <= upperlimit)]

newdfcap2 = df.copy()
newdfcap2['cgpa'] = np.where(
newdfcap2['cgpa'] > upperlimit, upperlimit ,
np.where(newdfcap2['cgpa'] < lowerlimit, lowerlimit, newdfcap2['cgpa']))
```

```python
plt.figure(figsize=(17,15))
plt.subplot(3,2,1)
sns.distplot(df['cgpa'])
    
plt.subplot(3,2,2)
sns.boxplot(df['cgpa'])

plt.subplot(3,2,3)
sns.distplot(newdf2['cgpa'])
    
plt.subplot(3,2,4)
sns.boxplot(newdf2['cgpa'])

plt.subplot(3,2,5)
sns.distplot(newdfcap2['cgpa'])

plt.subplot(3,2,6)
sns.boxplot(newdfcap2['cgpa'])
```
