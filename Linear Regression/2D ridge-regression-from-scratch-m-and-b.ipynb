```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2, random_state = 2)
```

```python
from sklearn.linear_model import Ridge, LinearRegression

r = Ridge(alpha = 10)
r.fit(xtrain,ytrain)
preds = r.predict(xtest)
```

```python
r2 = Ridge(alpha = 100)
r2.fit(xtrain, ytrain)
```

```python
lr = LinearRegression()
lr.fit(xtrain, ytrain)
```

```python
r.coef_, r.intercept_
```

```python
from sklearn.metrics import r2_score
r2_score(ytest, preds)
```

```python
# we can also compare the plots for different alpha values , higher alpha shows the line less towards the points 
plt.scatter(x,y)
plt.plot(x,lr.predict(x), c = 'black')
plt.plot(x,r.predict(x), c = 'red')
plt.plot(x,r2.predict(x), c = 'green')
```
