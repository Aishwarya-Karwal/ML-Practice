```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
```

```python
df = pd.read_csv(r"E:\Jupyter Notebooks\concrete_data.csv")
```

```python
df.head()
```

```python
df.isnull().sum() # luckily no null values :)
```

## Fit linear regression without power transform first

```python
xtrain, xtest, ytrain, ytest = train_test_split(df.iloc[:,:-1], df.iloc[:,-1], random_state = 42, 
                                               test_size = 0.2)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
lr = LinearRegression()

lr.fit(xtrain,ytrain)
pred = lr.predict(xtest)

r2_score(ytest, pred)
```

```python
# cross checking with cross val score
np.mean(cross_val_score(lr, xtrain, ytrain, cv = 10, scoring = 'r2'))
```

Distribution of the data without any transform

```python
for col in xtrain.columns:
    plt.figure(figsize = (14,5))
    plt.subplot(121)
    sns.distplot(xtrain[col])
    plt.title(col)
    
    plt.subplot(122)
    stats.probplot(xtrain[col], plot = plt, dist = "norm")
    plt.show()
```

## Again fitting the model after powertransformer

```python
# checking if the data is strictly positive to apply box-cox
xtrain.describe()
```

```python
ptf = PowerTransformer(method = "box-cox") # also by default standardizes the values
xtrainT=ptf.fit_transform(xtrain+0.0000001) 
# adding very small values to apply box-cox even on data with 0s
xtestT=ptf.transform(xtest+0.0000001)

pd.DataFrame({"cols":xtrain.columns, "lambdas":ptf.lambdas_})
```

```python
xtrainT # its a numpy arr , so we'll a df out of these values to plot them in next steps
```

```python
lr.fit(xtrainT, ytrain)
pred2 = lr.predict(xtestT)
r2_score(ytest, pred2)
```

```python
np.mean(cross_val_score(lr, xtrainT, ytrain, scoring = 'r2', cv = 10))
```

```python
dfboxcox = pd.DataFrame(xtrainT, columns = xtrain.columns)
```

```python
def plotdist(beforedf, afterdf):
    for col in beforedf.columns:
        plt.figure(figsize = (14,5))
        plt.subplot(121)
        sns.distplot(beforedf[col])
        plt.title("before {0}".format(col))
    
        plt.subplot(122)
        sns.distplot(afterdf[col])
        plt.title("after {0}".format(col))
        plt.show()
```

```python
plotdist(xtrain, dfboxcox)
```

```python
# now applying yeo-jhonson
# this is the default one

pt1 = PowerTransformer()

X_train_transformed2 = pt1.fit_transform(xtrain)
X_test_transformed2 = pt1.transform(xtest)

lr = LinearRegression()
lr.fit(X_train_transformed2,ytrain)

y_pred3 = lr.predict(X_test_transformed2)

print(r2_score(ytest,y_pred3))
pd.DataFrame({'cols':xtrain.columns,'Yeo_Johnson_lambdas':pt1.lambdas_})
```

```python
dfyeojohnson = pd.DataFrame(X_train_transformed2, columns = xtrain.columns)
```

```python
plotdist(xtrain, dfyeojohnson)
```

### side by side lambdas of box-cox and yeo-johnson

```python
pd.DataFrame({"cols":xtrain.columns, "box cox":ptf.lambdas_, "yeo johnson" : pt1.lambdas_})
```

### comparision of box cox and yeo johnson through probability distribution

```python
plotdist(dfboxcox, dfyeojohnson)
```
