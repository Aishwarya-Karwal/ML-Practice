```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer, MissingIndicator

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r"E:\Jupyter Notebooks\train.csv",usecols=['Age','Fare','Survived'])
```

```python
df.tail()
```

```python
df.isnull().mean()*100
```

### fitting model after imputing missing values with mean

```python
xtrain, xtest, ytrain, ytest = train_test_split(df.drop(['Survived'], axis=1),df['Survived']
                                               ,test_size=0.2, random_state=2)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
si = SimpleImputer()
xtraint = si.fit_transform(xtrain)
```

```python
xtraint
```

```python
xtestt = si.transform(xtest)
```

```python
clf = LogisticRegression()
clf.fit(xtraint, ytrain)
pred1 = clf.predict(xtestt)
```

```python
accuracy_score(ytest,pred1)
```

### fitting model after creating missing indicator column

```python
xtrain, xtest, ytrain, ytest = train_test_split(df.drop(['Survived'], axis=1),df['Survived']
                                               ,test_size=0.2, random_state=2)
```

### using simple imputer class itself to create missing indicator

```python
xtrain, xtest, ytrain, ytest = train_test_split(df.drop(['Survived'], axis=1),df['Survived']
                                               ,test_size=0.2, random_state=2)
```
